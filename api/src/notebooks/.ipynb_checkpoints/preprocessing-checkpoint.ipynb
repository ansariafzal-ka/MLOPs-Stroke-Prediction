## Pre-processing steps
- Imoute missing values in bmi feature with median
- Apply log transform to bmi and avg_glucose_level feature
- Remove "Other" category from gender feature
- Group rare categories: Never_wored and children

## Importing Libraries

```python
import pandas as pd
import numpy as np
```

## Configuations

```python
# path of the dataset
data_path='C:/Users/ansar/Desktop/Workspace/Personal/MLOPs/Storke Prediction/api/src/notebooks/data/stroke-data.csv'

# disable the limit for viewing all rows
pd.set_option('display.max_rows', None)
# disable the limit for viewing all columns
pd.set_option('display.max_columns', None)
```

## Pre-processing

```python
# loading the dataset
df = pd.read_csv(data_path)
```

```python
# dropping the missing values in bmi feature
df['bmi'] = df['bmi'].fillna(df['bmi'].median())
```

```python
# applying log transformation to bmi and avg_glucose_level features
df['bmi'] = np.log1p(df['bmi'])
df['avg_glucose_level'] = np.log1p(df['avg_glucose_level'])
```

```python
# remove the Other category from gender feature
df = df[df['gender'] != 'Other']
```

```python
# grouping rare work types
df['work_type'] = df['work_type'].replace({
    'Never_worked': 'Other',
    'children': 'Other'
})
```

```python
# saving the preprocessed data
df.to_csv('./data/preprocessed.csv', index=False, header=True)
print('preprocessed data saved successfully.')
```
