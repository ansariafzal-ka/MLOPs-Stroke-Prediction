## Importing Libraries

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
```

## Configurations

```python
# path of the dataset
data_path='C:/Users/ansar/Desktop/Workspace/Personal/MLOPs/Storke Prediction/api/src/notebooks/data/preprocessed.csv'

# disable the limit for viewing all rows
pd.set_option('display.max_rows', None)
# disable the limit for viewing all columns
pd.set_option('display.max_columns', None)
```

## Feature Engineering

```python
# loading the processed dataset
df_processed = pd.read_csv(data_path)
```

```python
# creating age threshold feature
df_processed['over_40'] = (df_processed['age'] > 40).astype(int)
```

```python
# creating medical risk score
df_processed['medical_risk_score'] = df_processed['hypertension'] + df_processed['heart_disease']
```

```python
# one-hot encoding gender and Residence_type features
df_processed = pd.get_dummies(df_processed, columns=['gender', 'Residence_type'], drop_first=True)
```

```python
# ordinal encoding ever_married, work_type, smoking_status features
df_processed['ever_married'] = df_processed['ever_married'].map({'Yes': 1, 'No': 0})
df_processed['work_type'] = df_processed['work_type'].map({'Self-employed': 2, 'Private': 1, 'Govt_job': 1, 'Other': 0})
df_processed['smoking_status'] = df_processed['smoking_status'].map({'formerly smoked': 2, 'smokes': 1, 'never smoked': 0, 'Unknown': 0})
```

```python
# scaling the features
continuous_features = ['age', 'bmi', 'avg_glucose_level']
scaler = StandardScaler()
df_processed[continuous_features] = scaler.fit_transform(df_processed[continuous_features])
```
